```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# reading csv file
cars=pd.read_csv('cars_sampled.csv')
cars
```

```python
cars.shape
```

```python
cars.info()
```

```python
# summarizing the data
cars.describe()
pd.set_option('display.float_format',lambda x:'%.3f'%x) # to display numbers upto 3 decimal points
cars.describe()
```

```python
# dropping unwanted columns
col=['name','dateCrawled','dateCreated','postalCode','lastSeen']
cars=cars.drop(columns=col,axis=1)
cars.head()
```

```python
cars.shape
```

```python
# removing duplicate records
cars.drop_duplicates(keep='first',inplace=True)
cars.shape
```

```python
cars.isnull().sum()
```

```python
# working range of data
cars=cars[
    (cars.yearOfRegistration<=2018)
  & (cars.yearOfRegistration>=1950)
  & (cars.price>=100)
  & (cars.price<=150000)
  & (cars.powerPS>=10)
  & (cars.powerPS<=500)]
cars.shape
```

```python
# creating new variable age by adding year & month of registration
cars['monthOfRegistration']/=12
cars['Age']=(2018-cars['yearOfRegistration'])+cars['monthOfRegistration']
cars['Age']=round(cars['Age'],2)
cars['Age'].describe()
```

```python
# dropping yearOfRegistration and monthOfRegistration
cars=cars.drop(columns=['yearOfRegistration','monthOfRegistration'],axis=1)
cars.shape
```

```python
# removing insignificant variables
col=['seller','offerType','abtest']
cars=cars.drop(columns=col,axis=1)
```

```python
# model building with imputed data
cars_imputed=cars.apply(lambda x:x.fillna(x.median())\
                       if x.dtype=='float'\
                       else x.fillna(x.value_counts().index[0]))
cars_imputed.isnull().sum()
```

```python
cars_imputed=pd.get_dummies(cars_imputed,drop_first=True)
cars_imputed.shape
```

```python
x=cars_imputed.drop(['price'],axis='columns',inplace=False)
y=cars_imputed['price']               
```

```python
# plotting variable price
prices=pd.DataFrame({"1. Before":y,"2. After":np.log(y)})
prices.hist()
```

```python
# transformming price as logarithmic value
y2=np.log(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y2,test_size=0.3,random_state=3)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
```

```python
# Linear Regression
lrg=LinearRegression(fit_intercept=True)
lin_mod=lrg.fit(x_train,y_train)
cars_predictions=lrg.predict(x_test)
```

```python
# R squared value
r2_train=lin_mod.score(x_test,y_test)
r2_test=lin_mod.score(x_train,y_train)
print(r2_train,r2_test)
```

```python
# Random Forest
rf=RandomForestRegressor(n_estimators=100,max_features='auto',max_depth=100,min_samples_split=10,min_samples_leaf=4,random_state=1)
```

```python
rf_model=rf.fit(x_train,y_train)
cars_predictions_rf=rf.predict(x_test)
```

```python
# R squared value
r2_rf_train=rf_model.score(x_train,y_train)
r2_rf_test=rf_model.score(x_test,y_test)
print(r2_rf_train,r2_rf_test)
```

```python
m=pd.Series(rf.feature_importances_,index=x_train.columns)
m.sort_values(ascending=False)
```
